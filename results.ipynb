```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_s = pd.DataFrame({
  'dim': [500, 1000, 1500, 2000, 2500, 3000],
  '2p': [1.42, 1.70, 1.62, 1.57, 1.67, 1.66],
  '4p': [2.3, 2.08, 1.89, 1.67, 1.73, 1.84]
})

df_e = pd.DataFrame({
  'dim': [500, 1000, 1500, 2000, 2500, 3000],
  '2p': [0.71, 0.85, 0.81, 0.79, 0.84, 0.83],
  '4p': [0.58, 0.52, 0.47, 0.42, 0.43, 0.46]
})
```

```python
plt.plot(df_s['dim'], df_s['2p'], label='2 процесори')
plt.plot(df_s['dim'], df_s['4p'], label='4 процесори')
plt.legend()
plt.ylabel('Прискорення')
plt.xlabel('Розмір матриці, n')
plt.show()
```

```python
plt.plot(df_e['dim'], df_e['2p'], label='2 процесори')
plt.plot(df_e['dim'], df_e['4p'], label='4 процесори')
plt.legend()
plt.ylabel('Ефективність')
plt.xlabel('Розмір матриці, n')
plt.show()
```
